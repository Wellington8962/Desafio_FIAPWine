### 3:

```python
# Removendo linhas duplicadas
data_wine2 = data_wine.drop_duplicates()
data_wine2
```

```python
# Removendo linhas com valores faltantes
data_wine3 = data_wine2.dropna()
data_wine3
```

### 4:

### a. DistribuiÃ§Ã£o da acidez volÃ¡til por qualidade do vinho.

```python
import matplotlib.pyplot as plt

# Calculando a mÃ©dia da acidez volÃ¡til (volatile acidity) para cada categoria de qualidade
media_volatile_acidity = data_wine3.groupby('quality')['volatile acidity'].mean()

# Criando um grÃ¡fico de barras
plt.figure(figsize=(8, 6))
bar_plot = plt.bar(media_volatile_acidity.index, media_volatile_acidity)
plt.xlabel('Qualidade do vinho')
plt.ylabel('MÃ©dia da acidez volÃ¡til')
plt.title('DistribuiÃ§Ã£o da Acidez VolÃ¡til por Qualidade do vinho')

# Adicionando os valores mÃ©dios nas barras
for bar, valor in zip(bar_plot, media_volatile_acidity):
    plt.annotate(f'{valor:.2f}', xy=(bar.get_x() + bar.get_width() / 2, valor), xytext=(0, 3), textcoords='offset points', ha='center')

plt.ylim(0, 1)
plt.xticks(media_volatile_acidity.index)
plt.show()
```

### b. DistribuiÃ§Ã£o do teor alcÃ³olico por qualidade do vinho.

```python
import matplotlib.pyplot as plt

# Calculando a mÃ©dia do teor alcoÃ³lico para cada categoria de qualidade
media_alcohol = data_wine3.groupby('quality')['alcohol'].mean()

# Criando um grÃ¡fico de barras
plt.figure(figsize=(8, 6))
bar_plot = plt.bar(media_alcohol.index, media_alcohol)
plt.xlabel('Qualidade do vinho', labelpad=10)
plt.ylabel('Teor AlcoÃ³lico MÃ©dio', labelpad=10)
plt.title('DistribuiÃ§Ã£o do Teor AlcoÃ³lico por Qualidade do Vinho')

# Adicionando os valores mÃ©dios nas barras
for bar, valor in zip(bar_plot, media_alcohol):
    plt.annotate(f'{valor:.2f}', xy=(bar.get_x() + bar.get_width() / 2, valor), xytext=(0, 3), textcoords='offset points', ha='center')

plt.ylim(0, 15)
plt.xticks(media_alcohol.index)
plt.show()
```

### c. RelaÃ§Ã£o entre acidez volÃ¡til e teor alcÃ³olico.

```python
# Extraindo as colunas de 'volatile acidity' e 'alcohol'
volatile_acidity = data_wine3['volatile acidity']
alcohol = data_wine3['alcohol']

# Criando um grÃ¡fico de dispersÃ£o(scatter plot)
plt.figure(figsize=(8, 6))
plt.scatter(volatile_acidity, alcohol, c='navy', alpha=0.5, label='Acidez volÃ¡til x Teor AlcÃ³olico')
plt.legend()
plt.xlabel('Acidez VolÃ¡til', labelpad=10)
plt.ylabel('Teor AlcÃ³olico', labelpad=10)
plt.title('RelaÃ§Ã£o entre Acidez VolÃ¡til e Teor AlcÃ³olico')
plt.grid(color='lightgray', linestyle='--', linewidth=0.5)
plt.show()
```

### d. RelaÃ§Ã£o entre teor alcÃ³olico e qualidade do vinho.

```python
# Criando um grÃ¡fico de dispersÃ£o (scatter plot)
plt.figure(figsize=(8, 6))
plt.scatter(data_wine3['alcohol'], data_wine3['quality'], c='blue', alpha=0.5)
plt.xlabel('Teor AlcÃ³olico', labelpad=10)
plt.ylabel('Qualidade do Vinho', labelpad=10)
plt.title('RelaÃ§Ã£o entre Teor AlcÃ³olico e Qualidade do Vinho')
plt.grid(color='lightgray', linestyle='--', linewidth=0.5)
plt.show()
```

### 5:

### a. Qual a acidez volÃ¡til mÃ©dia dos vinhos de qualidade elevada?

```python
# Filtrando as linhas com qualidade igual ou maior que 7
vinhos_alta_qualidade = data_wine3[data_wine3['quality'] >= 7]

# Calcular a mÃ©dia da acidez volÃ¡til para os vinhos de alta qualidade
media_acidez_volatil_alta_qualidade = vinhos_alta_qualidade['volatile acidity'].mean()

print(f"A mÃ©dia da acidez volÃ¡til dos vinhos de alta qualidade Ã©: {media_acidez_volatil_alta_qualidade:.2f}")
```

### b. Qual a relaÃ§Ã£o entre teor alcÃ³olico e qualidade do vinho?

Aqui estÃ£o algumas anÃ¡lises com base no grÃ¡fico:

**TendÃªncia Geral**: HÃ¡ uma tendÃªncia geral de que, Ã  medida que o teor alcÃ³olico aumenta, a qualidade do vinho tambÃ©m tende a aumentar. No entanto, essa relaÃ§Ã£o nÃ£o Ã© estritamente linear.

**CorrelaÃ§Ã£o Positiva**: Existe uma correlaÃ§Ã£o positiva entre o teor alcÃ³olico e a qualidade do vinho. Isso significa que, em mÃ©dia, vinhos com teor alcÃ³olico mais alto tendem a ter classificaÃ§Ãµes de qualidade mais altas.

**VariaÃ§Ã£o**: Apesar da tendÃªncia geral, ainda hÃ¡ uma quantidade significativa de dispersÃ£o nos dados. Isso significa que a qualidade do vinho nÃ£o Ã© determinada apenas pelo teor alcÃ³olico. Outros fatores tambÃ©m desempenham um papel na classificaÃ§Ã£o da qualidade.

**Outliers**: HÃ¡ alguns pontos que se destacam e nÃ£o seguem completamente a tendÃªncia. Alguns vinhos com teor alcÃ³olico moderado ainda tÃªm classificaÃ§Ãµes de alta qualidade, e vice-versa.

Abaixo segue o cÃ¡lculo do coeficiente de correlaÃ§Ã£o de Pearson, utilizando a biblioteca NumPy do Python:

```python
import numpy as np

# Dados de teor alcÃ³olico e qualidade do vinho
teor_alcoolico = data_wine3['alcohol']
qualidade_vinho = data_wine3['quality']

# Calculando o coeficiente de correlaÃ§Ã£o de Pearson utilizando a biblioteca NumPy
coeficiente_correlacao = np.corrcoef(teor_alcoolico, qualidade_vinho)[0, 1]

print(f"Coeficiente de CorrelaÃ§Ã£o de Pearson: {coeficiente_correlacao:.2f}")
```

Dado o resultado do coeficiente de CorrelaÃ§Ã£o de Pearson sendo igual a 0.48, pode-se concluir que Ã© um valor positivo e indica uma correlaÃ§Ã£o positiva, o que significa que, Ã  medida que o teor alcÃ³olico aumenta, a qualidade do vinho tende a aumentar.

### c. HÃ¡ alguma relaÃ§Ã£o entre acidez volÃ¡til e teor alcÃ³olico?

```python
# Calculando o coeficiente de correlaÃ§Ã£o de Pearson utilizando a biblioteca NumPy
acidez_volatil = data_wine3['volatile acidity']
coeficiente_correlacao = np.corrcoef(acidez_volatil, teor_alcoolico)[0, 1]

print(f"Coeficiente de CorrelaÃ§Ã£o de Pearson: {coeficiente_correlacao:.2f}")
```

Com base no coeficiente de CorrelaÃ§Ã£o de Pearson de -0.20, podemos concluir que a correlaÃ§Ã£o entre as variÃ¡veis Ã© relativamente fraca e prÃ³xima de zero. Isso sugere que nÃ£o hÃ¡ uma relaÃ§Ã£o linear significativa entre a acidez volÃ¡til e o teor alcÃ³olico dos vinhos analisados.

### d. HÃ¡ alguma relaÃ§Ã£o entre acidez volÃ¡til e qualidade do vinho?

```python
# Calculando o coeficiente de correlaÃ§Ã£o de Pearson utilizando a biblioteca NumPy
acidez_volatil = data_wine3['volatile acidity']
coeficiente_correlacao = np.corrcoef(acidez_volatil, qualidade_vinho)[0, 1]

print(f"Coeficiente de CorrelaÃ§Ã£o de Pearson: {coeficiente_correlacao:.2f}")
```
