### 3:

```python
# Removendo linhas duplicadas
data_wine2 = data_wine.drop_duplicates()
data_wine2
```

```python
# Removendo linhas com valores faltantes
data_wine3 = data_wine2.dropna()
data_wine3
```

### 4:

### a. Distribuição da acidez volátil por qualidade do vinho.

```python
import matplotlib.pyplot as plt

# Calculando a média da acidez volátil (volatile acidity) para cada categoria de qualidade
media_volatile_acidity = data_wine3.groupby('quality')['volatile acidity'].mean()

# Criando um gráfico de barras
plt.figure(figsize=(8, 6))
bar_plot = plt.bar(media_volatile_acidity.index, media_volatile_acidity)
plt.xlabel('Qualidade do vinho')
plt.ylabel('Média da acidez volátil')
plt.title('Distribuição da Acidez Volátil por Qualidade do vinho')

# Adicionando os valores médios nas barras
for bar, valor in zip(bar_plot, media_volatile_acidity):
    plt.annotate(f'{valor:.2f}', xy=(bar.get_x() + bar.get_width() / 2, valor), xytext=(0, 3), textcoords='offset points', ha='center')

plt.ylim(0, 1)
plt.xticks(media_volatile_acidity.index)
plt.show()
```

### b. Distribuição do teor alcóolico por qualidade do vinho.

```python
import matplotlib.pyplot as plt

# Calculando a média do teor alcoólico para cada categoria de qualidade
media_alcohol = data_wine3.groupby('quality')['alcohol'].mean()

# Criando um gráfico de barras
plt.figure(figsize=(8, 6))
bar_plot = plt.bar(media_alcohol.index, media_alcohol)
plt.xlabel('Qualidade do vinho', labelpad=10)
plt.ylabel('Teor Alcoólico Médio', labelpad=10)
plt.title('Distribuição do Teor Alcoólico por Qualidade do Vinho')

# Adicionando os valores médios nas barras
for bar, valor in zip(bar_plot, media_alcohol):
    plt.annotate(f'{valor:.2f}', xy=(bar.get_x() + bar.get_width() / 2, valor), xytext=(0, 3), textcoords='offset points', ha='center')

plt.ylim(0, 15)
plt.xticks(media_alcohol.index)
plt.show()
```

### c. Relação entre acidez volátil e teor alcóolico.

```python
# Extraindo as colunas de 'volatile acidity' e 'alcohol'
volatile_acidity = data_wine3['volatile acidity']
alcohol = data_wine3['alcohol']

# Criando um gráfico de dispersão(scatter plot)
plt.figure(figsize=(8, 6))
plt.scatter(volatile_acidity, alcohol, c='navy', alpha=0.5, label='Acidez volátil x Teor Alcóolico')
plt.legend()
plt.xlabel('Acidez Volátil', labelpad=10)
plt.ylabel('Teor Alcóolico', labelpad=10)
plt.title('Relação entre Acidez Volátil e Teor Alcóolico')
plt.grid(color='lightgray', linestyle='--', linewidth=0.5)
plt.show()
```

### d. Relação entre teor alcóolico e qualidade do vinho.

```python
# Criando um gráfico de dispersão (scatter plot)
plt.figure(figsize=(8, 6))
plt.scatter(data_wine3['alcohol'], data_wine3['quality'], c='blue', alpha=0.5)
plt.xlabel('Teor Alcóolico', labelpad=10)
plt.ylabel('Qualidade do Vinho', labelpad=10)
plt.title('Relação entre Teor Alcóolico e Qualidade do Vinho')
plt.grid(color='lightgray', linestyle='--', linewidth=0.5)
plt.show()
```

### 5:

### a. Qual a acidez volátil média dos vinhos de qualidade elevada?

```python
# Filtrando as linhas com qualidade igual ou maior que 7
vinhos_alta_qualidade = data_wine3[data_wine3['quality'] >= 7]

# Calcular a média da acidez volátil para os vinhos de alta qualidade
media_acidez_volatil_alta_qualidade = vinhos_alta_qualidade['volatile acidity'].mean()

print(f"A média da acidez volátil dos vinhos de alta qualidade é: {media_acidez_volatil_alta_qualidade:.2f}")
```

### b. Qual a relação entre teor alcóolico e qualidade do vinho?

Aqui estão algumas análises com base no gráfico:

**Tendência Geral**: Há uma tendência geral de que, à medida que o teor alcóolico aumenta, a qualidade do vinho também tende a aumentar. No entanto, essa relação não é estritamente linear.

**Correlação Positiva**: Existe uma correlação positiva entre o teor alcóolico e a qualidade do vinho. Isso significa que, em média, vinhos com teor alcóolico mais alto tendem a ter classificações de qualidade mais altas.

**Variação**: Apesar da tendência geral, ainda há uma quantidade significativa de dispersão nos dados. Isso significa que a qualidade do vinho não é determinada apenas pelo teor alcóolico. Outros fatores também desempenham um papel na classificação da qualidade.

**Outliers**: Há alguns pontos que se destacam e não seguem completamente a tendência. Alguns vinhos com teor alcóolico moderado ainda têm classificações de alta qualidade, e vice-versa.

Abaixo segue o cálculo do coeficiente de correlação de Pearson, utilizando a biblioteca NumPy do Python:

```python
import numpy as np

# Dados de teor alcóolico e qualidade do vinho
teor_alcoolico = data_wine3['alcohol']
qualidade_vinho = data_wine3['quality']

# Calculando o coeficiente de correlação de Pearson utilizando a biblioteca NumPy
coeficiente_correlacao = np.corrcoef(teor_alcoolico, qualidade_vinho)[0, 1]

print(f"Coeficiente de Correlação de Pearson: {coeficiente_correlacao:.2f}")
```

Dado o resultado do coeficiente de Correlação de Pearson sendo igual a 0.48, pode-se concluir que é um valor positivo e indica uma correlação positiva, o que significa que, à medida que o teor alcóolico aumenta, a qualidade do vinho tende a aumentar.

### c. Há alguma relação entre acidez volátil e teor alcóolico?

```python
# Calculando o coeficiente de correlação de Pearson utilizando a biblioteca NumPy
acidez_volatil = data_wine3['volatile acidity']
coeficiente_correlacao = np.corrcoef(acidez_volatil, teor_alcoolico)[0, 1]

print(f"Coeficiente de Correlação de Pearson: {coeficiente_correlacao:.2f}")
```

Com base no coeficiente de Correlação de Pearson de -0.20, podemos concluir que a correlação entre as variáveis é relativamente fraca e próxima de zero. Isso sugere que não há uma relação linear significativa entre a acidez volátil e o teor alcóolico dos vinhos analisados.

### d. Há alguma relação entre acidez volátil e qualidade do vinho?

```python
# Calculando o coeficiente de correlação de Pearson utilizando a biblioteca NumPy
acidez_volatil = data_wine3['volatile acidity']
coeficiente_correlacao = np.corrcoef(acidez_volatil, qualidade_vinho)[0, 1]

print(f"Coeficiente de Correlação de Pearson: {coeficiente_correlacao:.2f}")
```

```python
# Calculando o coeficiente de correlação de Pearson usando a função `corr` do Pandas
correlacao = data_wine3['volatile acidity'].corr(data_wine3['quality'])

print(f"Coeficiente de Correlação de Pearson: {correlacao:.2f}")
```
